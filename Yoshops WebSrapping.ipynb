```python
import urllib.request
from bs4 import BeautifulSoup
import csv
```

```python
url="https://yoshops.com/products?page=9"
r=urllib.request.Request(url)
#print(r)
with urllib.request.urlopen(r) as response:
    content=response.read()
    print(content)
```

```python
soup=BeautifulSoup(content,"html.parser")
print(soup)
```

```python
soup2=BeautifulSoup(soup.prettify(),"html.parser")
print(soup2)
```

```python
table=soup2.find("div",attrs={"class":"products"})
print(table.prettify)
```

```python
src="//onlinestore.wsimg.com/assets/noimage/product-5fec99477aebb10bac85d82665ec1497de4536cda3279e59089555c45cf589fa.png"
info=[]
for row in table.findAll("div",attrs={"class":"col-sm-3 col-xs-6"}):
    prod={}
    if row.img["src"]==src:
        prod["miss_img"]="www.yoshops.com"+row.img["src"]
        prod["product_url"]="www.yoshops.com"+row.a["href"]
        prod["product-title"]=row.span["href"].split("/")[2]
        prod["address"]=row.span["href"].split("-")[-1]
        info.append(prod)
print(prod)
print(info)
    
    
```

```python
filename="C:\\Users\\Dayanand\\Desktop\\DataScience\\dsp1\\yoshops_miss_products.csv"
with open(filename,"w",newline="") as f:
    w=csv.DictWriter(f,["miss_img","product_url","product-title","address"])
    w.writeheader()
    for i in info:
        w.writerow(i)
```
